# ðŸ”¥ LinkedIn Swipe: Professional Tinder Dataset Overview

## Welcome to the First LinkedIn Dataset with Red Flags & AI Conversation Starters!

**Problem:** 90% of LinkedIn connection requests are useless. You waste time networking with people who can't help you (and you can't help them).

**Solution:** This dataset contains compatibility scores for 500K professional pairs, plus:
- ðŸš¨ **Red Flag Detection** - Identify spam, collectors, job hoppers
- ðŸ’Ž **Hidden Gems** - Find undervalued high-potential connections
- ðŸ’¬ **AI Conversation Starters** - Personalized icebreakers
- ðŸ“Š **ROI Predictions** - Job opportunity, mentorship, collaboration scores

---

### Dataset Files:
1. `profiles_enhanced.csv` - 50,000 profiles with red flags & gems
2. `compatibility_pairs_enhanced.csv` - 500,000 pairs with ROI predictions
3. `conversation_starters.csv` - AI-generated icebreakers
4. `skills_network.csv` - Supply/demand analysis
5. `red_flags_top_profiles.csv` - Spam/problematic profiles
6. `hidden_gems_top_profiles.csv` - Undervalued opportunities

Let's explore! ðŸš€

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from collections import Counter

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("âœ… Libraries loaded!")
```

## ðŸ“Š Load the Data

```python
# Load profiles
profiles = pd.read_csv('../data/processed/profiles_enhanced.csv')
print(f"Loaded {len(profiles):,} profiles")

# Load compatibility pairs
pairs = pd.read_csv('../data/processed/compatibility_pairs_enhanced.csv')
print(f"Loaded {len(pairs):,} compatibility pairs")

# Load metadata
with open('../data/processed/metadata_enhanced.json', 'r') as f:
    metadata = json.load(f)

print("\nðŸ“ˆ Dataset Summary:")
for key, value in metadata.items():
    if isinstance(value, dict):
        continue
    print(f"  â€¢ {key}: {value}")
```

## ðŸ‘¥ Profile Demographics

```python
# Seniority distribution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Seniority
seniority_counts = profiles['seniority_level'].value_counts()
axes[0].bar(seniority_counts.index, seniority_counts.values, color='steelblue')
axes[0].set_title('Seniority Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Seniority Level')
axes[0].set_ylabel('Count')

# Industry
top_industries = profiles['industry'].value_counts().head(10)
axes[1].barh(top_industries.index, top_industries.values, color='coral')
axes[1].set_title('Top 10 Industries', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Count')
axes[1].invert_yaxis()

# Years of experience
axes[2].hist(profiles['years_experience'], bins=30, color='mediumseagreen', edgecolor='black')
axes[2].set_title('Years of Experience Distribution', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Years')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()

print(f"\nðŸ“Š Profile Stats:")
print(f"  â€¢ Avg Experience: {profiles['years_experience'].mean():.1f} years")
print(f"  â€¢ Avg Connections: {profiles['connections'].mean():.0f}")
print(f"  â€¢ Most Common Industry: {profiles['industry'].mode()[0]}")
```

## ðŸš¨ Red Flags Analysis

```python
# Red flag statistics
red_flag_stats = {
    'Connection Collectors': profiles['is_connection_collector'].sum(),
    'Job Hoppers': profiles['is_job_hopper'].sum(),
    'Ghost Profiles': profiles['is_ghost_profile'].sum(),
    'High Red Flag Score (>50)': (profiles['red_flag_score'] > 50).sum(),
    'Clean Profiles': (profiles['red_flag_score'] < 20).sum()
}

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Red flag types
axes[0].bar(red_flag_stats.keys(), red_flag_stats.values(), color=['red', 'orange', 'yellow', 'darkred', 'green'])
axes[0].set_title('ðŸš¨ Red Flag Profile Types', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Red flag score distribution
axes[1].hist(profiles['red_flag_score'], bins=50, color='crimson', alpha=0.7, edgecolor='black')
axes[1].axvline(50, color='black', linestyle='--', linewidth=2, label='High Risk Threshold')
axes[1].set_title('Red Flag Score Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Red Flag Score (0-100)')
axes[1].set_ylabel('Count')
axes[1].legend()

plt.tight_layout()
plt.show()

print("\nðŸš¨ Red Flag Summary:")
for flag, count in red_flag_stats.items():
    pct = (count / len(profiles)) * 100
    print(f"  â€¢ {flag}: {count:,} ({pct:.1f}%)")
```

## ðŸ’Ž Hidden Gems Analysis

```python
# Gem score distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gem score histogram
axes[0].hist(profiles['gem_score'], bins=50, color='gold', alpha=0.7, edgecolor='black')
axes[0].axvline(70, color='darkgreen', linestyle='--', linewidth=2, label='High Potential Threshold')
axes[0].set_title('ðŸ’Ž Hidden Gem Score Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Gem Score (0-100)')
axes[0].set_ylabel('Count')
axes[0].legend()

# Gem types
gem_types = profiles[profiles['gem_score'] > 40]['gem_type'].value_counts()
if len(gem_types) > 0:
    axes[1].pie(gem_types.values, labels=gem_types.index, autopct='%1.1f%%', startangle=90)
    axes[1].set_title('Gem Type Distribution', fontsize=14, fontweight='bold')
else:
    axes[1].text(0.5, 0.5, 'No High Gems Found', ha='center', va='center', fontsize=16)
    axes[1].set_xlim(0, 1)
    axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()

high_gems = (profiles['gem_score'] > 70).sum()
print(f"\nðŸ’Ž Found {high_gems:,} high-potential hidden gems (score > 70)")
print(f"  â€¢ Avg Gem Score: {profiles['gem_score'].mean():.1f}")
print(f"  â€¢ Top Gem Score: {profiles['gem_score'].max():.1f}")
```

## ðŸ¤ Compatibility Analysis

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Compatibility score distribution
axes[0, 0].hist(pairs['compatibility_score'], bins=50, color='skyblue', edgecolor='black')
axes[0, 0].axvline(pairs['compatibility_score'].mean(), color='red', linestyle='--', linewidth=2, label=f"Mean: {pairs['compatibility_score'].mean():.1f}")
axes[0, 0].set_title('ðŸ¤ Compatibility Score Distribution', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Compatibility Score (0-100)')
axes[0, 0].set_ylabel('Count')
axes[0, 0].legend()

# ROI Predictors
roi_cols = ['predicted_job_opportunity_score', 'predicted_mentorship_value', 'predicted_collaboration_potential']
roi_means = [pairs[col].mean() for col in roi_cols]
axes[0, 1].bar(['Job Opportunity', 'Mentorship', 'Collaboration'], roi_means, color=['purple', 'orange', 'green'])
axes[0, 1].set_title('ðŸ“Š Average ROI Predictions', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Score (0-100)')
axes[0, 1].set_ylim(0, 100)

# ROI Timeframe
timeframe_counts = pairs['roi_timeframe'].value_counts()
axes[1, 0].pie(timeframe_counts.values, labels=timeframe_counts.index, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('â±ï¸ Expected ROI Timeframe', fontsize=14, fontweight='bold')

# Skill complementarity vs compatibility
sample = pairs.sample(min(5000, len(pairs)))
axes[1, 1].scatter(sample['skill_complementarity_score'], sample['compatibility_score'], alpha=0.3, s=10)
axes[1, 1].set_title('Skills vs Compatibility', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Skill Complementarity')
axes[1, 1].set_ylabel('Overall Compatibility')

plt.tight_layout()
plt.show()

print("\nðŸ¤ Compatibility Insights:")
print(f"  â€¢ Avg Compatibility: {pairs['compatibility_score'].mean():.1f}/100")
print(f"  â€¢ High-Value Pairs (â‰¥70): {(pairs['compatibility_score'] >= 70).sum():,}")
print(f"  â€¢ Avg Job Opportunity Score: {pairs['predicted_job_opportunity_score'].mean():.1f}")
print(f"  â€¢ Avg Mentorship Value: {pairs['predicted_mentorship_value'].mean():.1f}")
```

## ðŸŽ¯ Key Insights

### What Makes This Dataset Unique?

1. **ðŸš¨ Red Flag Detection** - First dataset to identify problematic LinkedIn profiles
   - Connection collectors who accept everyone
   - Job hoppers (flight risk)
   - Ghost profiles with minimal info
   - Spam likelihood scores

2. **ðŸ’Ž Hidden Gems** - Identify undervalued opportunities
   - Rising stars (fast career growth)
   - Super connectors in niche areas
   - Rare, in-demand skills

3. **ðŸ“Š ROI Predictions** - Not just compatibility scores
   - Job opportunity likelihood
   - Mentorship value
   - Collaboration potential
   - Expected timeframe for value

4. **ðŸ’¬ Actionable Insights** - Detailed explanations with specific advice
   - What they can help you with
   - What you can offer them
   - Specific action items

- Career path optimization
